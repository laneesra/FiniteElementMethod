```python
data = '9,5710,0710,7710,249,989,659,3010,3311,519,2310,329,1210,339,2810,5710,2410,6210,1810,8511,029,7810,4210,9010,239,4510,5010,4811,119,5310,0511,589,7210,599,6810,929,8710,2710,2210,9710,8210,6610,6910,809,4210,6910,5410,8510,2410,4810,3511,079,5411,189,6711,439,8010,8611,2510,2310,089,7511,0510,0710,0310,5710,279,979,9210,6210,8710,4710,1210,089,999,969,859,8510,6310.229,309,8310,7510,6510,209,579,8910,1710,0510,0210,3510,3410,229,7510,009,8510,7711,2310,0510,3010,0310,739,7910,8810,0310,1710,229,1010,0211,5311.409,809,809,8310,1310,2310,5011,4510,5110,6710,4810,779,9710,7210,5510,4211,669,319,4610,0011,359,3310,0510,2710,3810,2410,4310,3011,6110,229,0810,3410,4111,2211,289,859,6310,0310,4010,9310,46'
data = data.replace('.', ',')
commas = [0] + [i + 3 for i in range(len(data)) if data[i] == ',']
values = [float(data[commas[i]:commas[i+1]].replace(',', '.')) for i in range(len(commas) - 1)]
print(values)
n = len(values)
print(n)
from collections import Counter
stat_series = Counter()

for value in values:
    stat_series[value] += 1

sorted_keys = sorted(stat_series.keys())

print('Here is statistic series with len {}:'.format(len(sorted_keys)))
for key in sorted_keys:
    print('key {key}, count {count}'.format(key=key, count=stat_series[key]))

print()


from math import log, floor
m = floor(log(n, 2)) + 1
print('m is ', m)

min_arg = sorted_keys[0]
max_arg = sorted_keys[-1]

print('min value is ', min_arg, ', max value is ', max_arg)

step = (max_arg - min_arg) / m

print('step is ', step)

def find_subslice_sum(left_border, right_border = None):

    def is_in_border(x):
        return x >= left_border and (right_border is None or x < right_border)

    matched_keys = list(filter(is_in_border, sorted_keys))
    return sum([stat_series[key] for key in matched_keys])

x_values_draw = []
y_values_draw = []

from numpy import arange
print('Here is interval statistic series:')
for slice_start in arange(min_arg, max_arg, step):
    if slice_start + step >= max_arg:
        slice_end = None
    else:
        slice_end = slice_start + step
    range_sum = find_subslice_sum(slice_start, slice_end)
    histo_value = range_sum / (n * step)

    x_values_draw.append(slice_start + step/2)
    y_values_draw.append(histo_value)

    if slice_end is None:
        print('[{start},{end}] : sum {sum}'.format(start=slice_start, end=max_arg, sum=range_sum))
    else:
        print('[{start},{end}) : sum {sum}'.format(start=slice_start, end=slice_end, sum=range_sum))

print()
from numpy import mean, var
print('Here is mean of statistic series: ', mean(values))
print()
print('Here is variance of statistic series: ', var(values, ddof=1))
print()


import matplotlib
import matplotlib.pyplot as plt
a,b,c = matplotlib.pyplot.hist(values, m)
import numpy as np
np.sum(a)
print(a)
print(b)
```
